## Analyze physical activity and high blood pressure

### Subtask:
Investigate the relationship between physical activity ('exercicio_físico') and high blood pressure ('pressao_alta'). This could involve grouping by both columns and observing the counts.


**Reasoning**:
Group the data by physical activity and blood pressure and count the occurrences to investigate the relationship.



```python
import matplotlib.pyplot as plt

# Paleta de cores claras
cores_claras = [
    '#A2D9FF', '#A3D9A5', '#D1D8E0', '#FFA07A', '#D9A2FF',
    '#FFD9A2', '#A2F0FF', '#FFDAA2', '#A2FFD9', '#D1FFDA'
]

# Agrupar os dados por faixa etária e pressão alta
blood_pressure_by_age = df_limpo.groupby(['faixa_etaria', 'pressao_alta']).size().unstack(fill_value=0)

# Separar os dados: 1 = tem pressão alta, 2 = não tem
com_pressao = blood_pressure_by_age[1] if 1 in blood_pressure_by_age.columns else None
sem_pressao = blood_pressure_by_age[2] if 2 in blood_pressure_by_age.columns else None

# Criar subplots lado a lado
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Gráfico 1: com pressão alta
if com_pressao is not None:
    axes[0].bar(com_pressao.index, com_pressao.values, color=cores_claras[:len(com_pressao)])
    axes[0].set_title('Faixas Etárias com Pressão Alta')
    axes[0].set_xlabel('Faixa Etária')
    axes[0].set_ylabel('Número de Pessoas')
    axes[0].tick_params(axis='x', rotation=45)

    # Adicionar valores sobre as barras
    for i, valor in enumerate(com_pressao.values):
        axes[0].text(i, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)

# Gráfico 2: sem pressão alta
if sem_pressao is not None:
    axes[1].bar(sem_pressao.index, sem_pressao.values, color=cores_claras[:len(sem_pressao)])
    axes[1].set_title('Faixas Etárias sem Pressão Alta')
    axes[1].set_xlabel('Faixa Etária')
    axes[1].tick_params(axis='x', rotation=45)

    # Adicionar valores sobre as barras
    for i, valor in enumerate(sem_pressao.values):
        axes[1].text(i, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)

# Ajustar layout
plt.tight_layout()
plt.savefig('faixa_etaria_com_pressao.png', dpi=300)
plt.show()
```
